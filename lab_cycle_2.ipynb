```python
def rabbit_population(n):

    pairs = [1, 1]
    for month in range(2, n):

        new_pairs = pairs[month - 1] + pairs[month - 2]
        pairs.append(new_pairs)

    return pairs

def print_rabbit_table(pairs):
    print("Month\tPairs of Rabbits")
    for month, pair_count in enumerate(pairs):
        print(f"{month + 1}\t{pair_count}")
N = 12


rabbit_pairs = rabbit_population(N)


print_rabbit_table(rabbit_pairs)
```

```python
def main():

    input_str = input("Enter a string of numbers separated by a space: ")
    numbers = [int(num) for num in input_str.split()]


    k = int(input("Enter the number of positions to rotate to the right: "))
    rotated_list = numbers[-k:] + numbers[:-k]


    tuple_from_list = tuple([num for num in rotated_list])


    unique_list = list(set(tuple_from_list))


    function_result_list = [x ** 2 - x for x in unique_list]


    sorted_function_result_list = sorted(function_result_list)
    sorted_unique_list = sorted(unique_list)
    merged_sorted_list = sorted_function_result_list + sorted_unique_list


    print("\nResults:")
    print("Rotated List:", rotated_list)
    print("Tuple from List:", tuple_from_list)
    print("Unique List:", unique_list)
    print("Function Result List:", function_result_list)
    print("Sorted Merged List:", merged_sorted_list)

if __name__ == "__main__":
    main()
```

```python
import json

#create a list with elements from file

file = open("iris.json", "r")
txt_file = file.read()
print(txt_file)

def totalA(flower):
    sepal_area = flower.get("sepalLength") * flower.get("sepalWidth")
    petal_area = flower.get("petalLength") * flower.get("petalWidth")
    return sepal_area + petal_area

# creating a list of dictionary objects

l_o_d = json.loads(txt_file)

# details of the flower species naming setosta

for i in l_o_d:
    if i.get("species") == "setosa":
        print(i)

petal_lst = []
sepal_lst = []
for j in l_o_d:
    sepal_length = j.get('sepalLength')
    sepal_width = j.get('sepalWidth')
    petal_length = j.get('petalLength')
    petal_width = j.get('petalWidth')

    # Check for None
    if all(v is not None for v in [sepal_length, sepal_width, petal_length, petal_width]):
        petal_area = round(petal_length * petal_width, 2)
        petal_lst.append(petal_area)
        sepal_area = round(sepal_length * sepal_width, 2)
        sepal_lst.append(sepal_area)
    else:
        print("Missing values for some attributes in the data.")

# Print the minimum petal area and max sepal area in each species

print("MAXIMUM SEPAL AREA:", max(sepal_lst))
print("MINIMUM PETAL AREA:", min(petal_lst))

#Sort the list of dictionaries according to the total area for sepal and petal

sorted_list = sorted(l_o_d, key=totalA, reverse=True)
print("Sorted List based on Total Area:")
for item in sorted_list:
    print(item)
```

```python
class box:
    def __init__(self,length,breadth=None,height=None):
        #rectangular prism
        self.Length=length
        self.Breadth=breadth
        self.Height=height

        #cube
        if breadth is None and height is None:
            self.Breadth=length
            self.Height=length
        elif height is None:
            self.Height=length

    def Area(self):
        area=2*(self.Length*self.Breadth+self.Breadth*self.Height+self.Length*self.Height)
        return area

    def volume(self):
        volu=self.Length*self.Breadth*self.Height
        return volu

cube=box(7)
sq_prism=box(7,3)
rect_prism=box(7,3,3)
print("AREA OF CUBE:", cube.Area())
print("VOLUME OF CUBE:",cube.volume())

print("AREA OF SQUARE PRISM:", sq_prism.Area())
print("VOLUME OF SQUARE PRISM:",sq_prism.volume())

print("AREA OF RECTANGULAR PRISM:",rect_prism.Area())
print("VOLUME OF RECTANGULAR PRISM:",rect_prism.volume())
```

```python
class _3Dshape:
    def print_volume(self):
        pass  # To be implemented in child classes

    def print_area(self):
        pass  # To be implemented in child classes

class cylinder(_3Dshape):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def print_volume(self):
        volume = 3.14159 * self.radius**2 * self.height
        print(f"Volume of cylinder: {volume:.2f} cubic units")

    def print_area(self):
        area = 2 * 3.14159 * self.radius * (self.radius + self.height)
        print(f"Surface Area of cylinder: {area:.2f} square units")

class Sphere(_3Dshape):
    def __init__(self, radius):
        self.radius = radius

    def print_volume(self):
        volume = (4/3) * 3.14159 * self.radius**3
        print(f"Volume of Sphere: {volume:.2f} cubic units")

    def print_area(self):
        area = 4 * 3.14159 * self.radius**2
        print(f"Surface Area of Sphere: {area:.2f} square units")

if __name__ == "__main__":

    cylinder = cylinder(radius=6, height=8)
    cylinder.print_volume()
    cylinder.print_area()

    print()

    sphere = Sphere(radius=3)
    sphere.print_volume()
    sphere.print_area()
```
